# Effectively Using Matplotlib

## Imports

````python
# Pandas has his own built in fixed functions but we need to use one custome called xlrd so please use before starting
# pip install xlrd
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
````

````python
df = pd.read_excel("https://github.com/chris1610/pbpython/blob/master/data/sample-salesv3.xlsx?raw=true")
df.head()
````

## Top 10 customers

````python
top10 = (df.groupby('name')['ext price', 'quantity'].agg({'ext price' : 'sum', 'quantity' : 'count'})
         .sort_values(by='ext price', ascending = False))[:10].reset_index()
top10
````

````python
# Above example labels does not look nice so lets change it a bit.
top_10 = (df.groupby('name')['ext price', 'quantity'].agg({'ext price': 'sum', 'quantity': 'count'})
          .sort_values(by='ext price', ascending=False))[:10].reset_index()

top_10.rename(columns={'name': 'Name', 'ext price': 'Sales', 'quantity': 'Purchases'}, inplace=True)
top_10
````

## Matplotlib

````python
# What are avaiable styles for rendering plots?
plt.style.available
````

````python
# You can experiment in here and choose the best one for you
plt.style.use('ggplot')
````

````python
# But now let's use standard pandas ploting
top_10.plot(kind='barh', y='Sales', x='Name')

plt.show()
````

Below you have list of **plot kind**s: 
* ‘line’ : line plot (default)
* ‘bar’ : vertical bar plot
* ‘barh’ : horizontal bar plot
* ‘hist’ : histogram
* ‘box’ : boxplot
* ‘kde’ : Kernel Density Estimation plot
* ‘density’ : same as ‘kde’
* ‘area’ : area plot
* ‘pie’ : pie plot
* ‘scatter’ : scatter plot
* ‘hexbin’ : hexbin plot

## Customizing the plot

````python
# It is critical to get access to axes and figures in matplotlib
fig, ax = plt.subplots()
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)

plt.show()
````

### You may ask why we want to have access to axes or figures?

````python
# Let's play around axes and set limits to it and labes
fig, ax = plt.subplots()
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)

ax.set_xlim([-20000, 140000])
ax.set_xlabel('Total Revenue')
ax.set_ylabel('Customer')

plt.show()
````

````python
# Here is the shorter verion of it
fig, ax = plt.subplots()
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)

ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel='Customer')

plt.show()
````

````python
# Going further we can change the size of the plot or disable the legend which is unnecessary in this case
fig, ax = plt.subplots(figsize=(5, 3))

top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel='Customer')

ax.legend().set_visible(False)

plt.show()
````

### Adding currency
```FuncFormatter``` can apply user defined funtion to the value abd return nicely formatter string to place on the axis

````python
# First of all we need to define function that will be changing the currencies
# We go either in milions or thousands
def currency(x, pos):
    'The two args are the value and tick position'
    if x == 0:
        return 0
    if x >= 1000000:
        return '${:1.1f}M'.format(x*1e-6)
    return '${:1.0f}K'.format(x*1e-3)
````

````python
# Now let's print this
fig, ax = plt.subplots()
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel='Customer')

formatter = FuncFormatter(currency)
ax.xaxis.set_major_formatter(formatter)
ax.legend().set_visible(False)

plt.show()
````

### Adding lines and custom text to the graph

````python
# Create the figure and the axes
fig, ax = plt.subplots()

# Plot the data and get the averaged
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax)
avg = top_10['Sales'].mean()
print('Top 10 customers mean value is: ',avg)

# Set limits and labels
ax.set_xlim([-10000, 140000])
ax.set(title='2014 Revenue', xlabel='Total Revenue', ylabel='Customer')

# Add a line for the average
ax.axvline(x=avg, color='b', label='Average', linestyle='--', linewidth=1)

# Annotate the new customers
new_customers = [1, 5 ,9]
for cust in new_customers:
    ax.text(125000, cust, "New Customer")

# Format the currency
formatter = FuncFormatter(currency)
ax.xaxis.set_major_formatter(formatter)

# Hide the legend
ax.legend().set_visible(False)

plt.show()
````

## Subplots

````python
# I used nrows and ncols but normally people just specify the value without the name parameter
# I am using sharey=True so that the yaxis will share the same labels for both figures.
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(7, 4))
````

````python
top_10.plot(kind='barh', y="Sales", x="Name", ax=ax0)
ax0.set_xlim([-10000, 140000])
ax0.set(title='Revenue', xlabel='Total Revenue', ylabel='Customers')

# Plot the average as a vertical line
avg = top_10['Sales'].mean()
ax0.axvline(x=avg, color='b', label='Average', linestyle='--', linewidth=1)

# Repeat for the unit plot
top_10.plot(kind='barh', y="Purchases", x="Name", ax=ax1)
avg = top_10['Purchases'].mean()
ax1.set(title='Units', xlabel='Total Units', ylabel='')
ax1.axvline(x=avg, color='b', label='Average', linestyle='--', linewidth=1)

# Title the figure
fig.suptitle('2014 Sales Analysis', fontsize=14, fontweight='bold');

# Hide the legends
ax1.legend().set_visible(False)
ax0.legend().set_visible(False)

plt.show()
````
